```python
import pandas as pd
import numpy as np
```

```python
# extraction of tables from website then creating 1 df for each table 
dfs = pd.read_html("https://www.mariowiki.com/Mario_Kart_8_Deluxe_in-game_statistics")
perso_df = dfs[0]
kart_df = dfs[1]
wheels_df = dfs[2]
glider_df = dfs[3]
```

```python
# Droping first mutli level column to keep only the short name
perso_df.columns = perso_df.columns.droplevel(0)
kart_df.columns = kart_df.columns.droplevel(0)
wheels_df.columns = wheels_df.columns.droplevel(0)
glider_df.columns = glider_df.columns.droplevel(0)
```

```python
# dropin last usless column from perso_df
perso_df = perso_df.iloc[:-1 , :]
```

```python
# converting stats columns from str to int
col_list = ['WG', 'AC', 'ON', 'OF', 'MT', 'SL', 'SW', 'SA', 'SG', 'TL',
       'TW', 'TA', 'TG', 'IV']
perso_df[col_list] = perso_df[col_list].astype(int)
kart_df[col_list] = kart_df[col_list].astype(int)
wheels_df[col_list] = wheels_df[col_list].astype(int)
glider_df[col_list] = glider_df[col_list].astype(int)
```

```python
# Computing stats series
stats = (perso_df.iloc[0,1:] + kart_df.iloc[0,1:] + wheels_df.iloc[0,1:] + glider_df.iloc[0,1:] + 3 )/ 4
```

```python
# Computing stats series
stats1 = (perso_df.iloc[1,1:] + kart_df.iloc[1,1:] + wheels_df.iloc[1,1:] + glider_df.iloc[1,1:] + 3 )/ 4
```

```python
# creating items series to concatenate to stats 
items_data = {'perso':perso_df['Driver'].loc[0],
        'kart':kart_df['Body'].loc[0], 
        'wheels':wheels_df['Tire'].loc[0], 
        'glider': glider_df['Glider'].loc[0]}
items = pd.Series(data=items_data, index=['perso', 'kart', 'wheels', 'glider'])
items
```

```python
# creating items series to concatenate to stats 
items_data1 = {'perso':perso_df['Driver'].loc[1],
        'kart':kart_df['Body'].loc[1], 
        'wheels':wheels_df['Tire'].loc[1], 
        'glider': glider_df['Glider'].loc[1]}
items1 = pd.Series(data=items_data1, index=['perso', 'kart', 'wheels', 'glider'])
items1
```

```python
# concatenate items + stats series
item_stats_df = pd.concat([items,stats])
item_stats_df 
```

```python
# concatenate items + stats series
item_stats_df1 = pd.concat([items1,stats1])
item_stats_df1 
```

```python
# creating function to compute all possible items-stats combinations (n=720k)

item_stats_df_vide = pd.DataFrame()

for p in range(52):
    for k in range(41):
        for w in range(22):
            for g in range(15):
    
                # Computing stats series
                stats = (perso_df.iloc[p,1:] + kart_df.iloc[k,1:] + wheels_df.iloc[w,1:] + glider_df.iloc[g,1:] + 3 )/ 4

                # creating items series to concatenate to stats 
                items_data = {'perso':perso_df['Driver'].loc[0],
                        'kart':kart_df['Body'].loc[0], 
                        'wheels':wheels_df['Tire'].loc[0], 
                        'glider': glider_df['Glider'].loc[0]}
                items = pd.Series(data=items_data, index=['perso', 'kart', 'wheels', 'glider'])

                # concatenate items + stats series
                item_stats_df = pd.concat([items,stats])
                
                # adding df to item_stats_df_complet
                item_stats_df_complet = item_stats_df_vide.append(item_stats_df, ignore_index=True)
                
                
                
```

```python
# creating function to compute all possible items-stats combinations (n=720k)

item_stats_df_complet = pd.DataFrame()

for p in range(52):
    for k in range(41):
        for w in range(22):
            for g in range(15):
    
                # Computing stats series
                stats = (perso_df.iloc[p,1:] + kart_df.iloc[k,1:] + wheels_df.iloc[w,1:] + glider_df.iloc[g,1:] + 3 )/ 4

                # creating items series to concatenate to stats 
                items_data = {'perso':perso_df['Driver'].loc[p],
                        'kart':kart_df['Body'].loc[k], 
                        'wheels':wheels_df['Tire'].loc[w], 
                        'glider': glider_df['Glider'].loc[g]}
                items = pd.Series(data=items_data, index=['perso', 'kart', 'wheels', 'glider']) 
                
                # concatenate items + stats series
                item_stats_series = pd.concat([items,stats])
                
                # coverting series to dict before concatenating
                item_stats_dict = item_stats_series.to_dict()
                
                # adding df to item_stats_df_complet
                item_stats_df_complet = pd.concat([item_stats_df_complet,
                                                   pd.DataFrame([item_stats_dict])],
                                                  ignore_index=True)                
                
                
```
